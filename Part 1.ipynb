#### 4. Check if there are any duplicate records in the dataset? if any drop them.

```python
duplicates = data.duplicated()
print("No of duplicate records", duplicates.sum())

data = data.drop_duplicates()
print("No of duplicate records after dropping ", data.duplicated().sum())
```

#### 5.  Check the unique categories of the column 'Category', Is there any invalid category? If yes, drop them.


```python
unique_categories = data['Category'].unique()
print("Unique categories \n",unique_categories)

invalid_categories = ['Invalid_Category1','Invalid_Category2']

data = data[~data['Category'].isin(invalid_categories)]

updated_unique_categories = data['Category'].unique()
print("unique categories after droping invalid category \n", updated_unique_categories)
```

#### 6. Check if there are missing values present in the column Rating, If any? drop them and and create a new column as 'Rating_category' by converting ratings to high and low categories(>3.5 is high rest low)

```python
missing_values = data['Rating'].isnull().sum()
print("Number of missing values in 'Rating' column:",missing_values)

data = data.dropna(subset=['Rating'])

data['Rating_category'] = ['high' if rating > 3.5 else 'low' for rating in data['Rating']]
# we created a new column Rating_category
print(data.head())
```

#### 8. Convert the column "Reviews'' to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach.(Hint: Use log transformation)

```python
data['Reviews'] = pd.to_numeric(data['Reviews'], errors = 'coerce')

Q1 = data['Reviews'].quantile(0.25)
Q3 = data['Reviews'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['Reviews'] < lower_bound) | (data['Reviews'] > upper_bound)]
print("Number of outlier:", len(outliers))

data['Reviews'] = np.log1p(data['Reviews'])

print(data.head())
```

#### 9. The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device')

```python
data['Size'] = data['Size'].replace({'M': 'e+6', 'K': 'e+3'}, regex = True)

data['Size'] = pd.to_numeric(data['Size'], errors = 'coerce')

data = data[data['Size'] != 'Varies with device']

print(data.head())
```

#### 10. Check the column 'Installs', treat the unwanted characters and convert the column into a suitable data type.

```python
data['Installs'] = data['Installs'].str.replace('[+,]', '' , regex = True)

data['Installs'] = pd.to_numeric(data['Installs'], errors = 'coerce')

print(data.head())
```

#### 11. Check the column 'Price' , remove the unwanted characters and convert the column into a suitable data type.

```python
data['Price'] = data['Price'].str.replace('$', '', regex = False)

data['Price'] = pd.to_numeric(data['Price'], errors = 'coerce')

print(data.head())
```
