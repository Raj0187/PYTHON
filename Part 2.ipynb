#### 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

#### 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')


```python
columns_using2 = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
using_data2 = data[columns_using2]

correlation_matrix = using_data2.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('correlation matrix')
plt.show()
```

```python
# You can observe the correlation values between pairs of variables, which helps understand the relationships and dependencies between them.

# Positive correlations (values closer to 1) are indicated by darker shades, while negative correlations (values closer to -1) are indicated by lighter shades.

# Variables with higher positive correlation values are likely to have a strong positive relationship, and variables with higher negative correlation values are likely to have a strong negative relationship.

# By examining the correlation values, you can identify variables that are strongly correlated with each other, indicating potential multicollinearity.

# Variables with correlation values close to zero indicate a weak or no linear relationship.
```
